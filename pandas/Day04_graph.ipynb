```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 직선그리기 : 시계열 데이터 표시할 때 사용하는 그래프
# 데이터 연동
x = np.linspace(0, 10 ,100)  # 가로 측 범위를 지정
# y = 2 * x + 3  # y = 2x+3
y = np.sin(x)  # 세로 측
y2 = np.cos(x)

# 차트 표현 변경
plt.ylim(-1.5, 1.5)  # (1) Y축의 표현 범위
plt.title("sin(x)" ,fontsize = 24)  # (2) 타이틀
plt.xlabel("X", fontsize = 24)  # (3) x레이블
plt.ylabel("Y", fontsize = 24)  # (4) y레이블
plt.grid(True)
plt.tick_params(labelsize = 13)  # (5) 눈금선 레이블 크기
plt.legend(loc='lower left', fontsize=10)  # (6) 범례 표시

# 차트그리기
# plt.plot(x,y, color = 'r')  # 선 그래프
plt.plot(x,y, color = (1.0, 0.0, 0.0), label = 'sin(x)')  # 선 그래프
plt.plot(x,y2, color = (0.0, 0.0, 1.0), label = 'cos(x)')  # 선 그래프
plt.show()
```

```python
# 막대 그래프
x = np.linspace(0, 10 ,10)  # 가로 측 범위를 지정
# y = 2 * x + 3  # y = 2x+3
y = np.sin(x)  # 세로 측
y2 = np.cos(x)

# 차트 표현 변경
plt.ylim(-1.5, 1.5)  # (1) Y축의 표현 범위
plt.title("sin(x)" ,fontsize = 24)  # (2) 타이틀
plt.xlabel("X", fontsize = 24)  # (3) x레이블
plt.ylabel("Y", fontsize = 24)  # (4) y레이블
plt.grid(True)
plt.tick_params(labelsize = 13)  # (5) 눈금선 레이블 크기
plt.legend(loc='lower left', fontsize=10)  # (6) 범례 표시

# 차트그리기
# plt.plot(x,y, color = 'r')  # 선 그래프
plt.bar(x,y, color = (1.0, 0.0, 0.0), label = 'sin(x)')  # 막대 그래프
plt.bar(x,y2, color = (0.0, 0.0, 1.0), label = 'cos(x)')  # 막대 그래프
plt.show()
```

```python
# 2. 막대 그래프 = csv 파일 호출
import pandas as pd
df_sales = pd.read_csv("sample_p.csv")
df_sales

plt.bar(df_sales['Product'], df_sales['Sales'], tick_label = df_sales['Product'], align='center', color='c')
plt.show
```

```python
# 3. 원형 그래프
df_sales = pd.read_csv("sample_p.csv")
df_sales

m_colors = ['r','c','b','m','y']
plt.pie(df_sales['Sales'], labels = df_sales['Product'], autopct = "%1.1f%%", colors = m_colors, counterclock = True)
plt.show
```

```python
# 4. 산점도(scatter) : 변수간의 상관 관계를 시각적으로 확인, 데이터의 편차를 통해서 값의 범위를 시각적으로 확인
# plt.scatter()
# california_housing_train.csv를 이용해서 그래프를 그려보자
df = pd.read_csv('california_housing_train.csv')
df.head(5)

# median_income의 열데이터와 median_house_value 있는 데이터 열을 각각 가로축, 세로축에 대한 산점도를 그려보자
plt.scatter(df['median_income'], df['median_house_value'])
```

```python
plt.scatter(df['population'], df['median_house_value'])
```

```python
# 히스토그램(histogram) : 데이터에 어떤 값이 자주 등장하고 있는지를 판별하기 위해 각 값의 빈도를 확인하기 위해 사용되는 것
# 히스토그램의 특징 : 막대 하나를 bin이라고 부른다(각각의 높이값의 범위에 들어있는 샘플의 수)
# plt.hist()
# df['median_house_value']의 히스토그램을 그려보자
plt.hist(df['median_house_value'], bins=50)
```

```python
# 상자 그래프 (boxplot) : 값의 차이를 표시하기 위한 그래프
# df.describe()
plt.figure()
plt.boxplot(df['median_house_value'])
plt.savefig('mybox.png')
plt.close('all')

plt.boxplot((df['total_bedrooms'], df['population']))
```

```python
import seaborn as sns

arr_2d = np.arange(-8, 8).reshape(4,4)
arr_2d
df = pd.DataFrame(data=arr_2d,  index=['a','b','c','d'], columns=['A','B','C','D'])

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

sns.heatmap(df, annot = True, cbar=True, vmax=10, vmin=-10, center=0)  # 최대값, 최소값, 중앙값
fig.savefig('myhit.png')

# 위치를 분할해서 원하는 위치에 heatmap을 그려낼 수 있다.
fig, axex = plt.subplots(nrows=2, ncols=3, figsize=(8,6))
sns.heatmap(df, ax = axex[0,0])
sns.heatmap(df, ax = axex[1,0])
```

```python
# 다수의 특징 량을 가지는 데이터의 상관 계수를 시각화해라
# 단 캐글의 주택 가격을 추정하는 문제의 학습 데이터를 사용해라
df_house = pd.read_csv('train.csv', index_col=0)
df_house

# 상관관계 dataFrame의 각 열사이의 상관계수를 산출 corr()
df_house_corr = df_house.corr()  # nan 값은 배제하고 산출
df_house_corr

# shape를 확인
print(df_house_corr.shape)

# 5개의 header값만 보자
df_house_corr.head(5)

fig, axex = plt.subplots(figsize=(12, 9))
sns.heatmap(df_house_corr, vmax=1, vmin=-1, center=0, square=True, cmap='Blues')
```

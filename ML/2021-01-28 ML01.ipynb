머신러닝의 분류작업에서 모델을 평가할 때 사용하는 그래프
임의로 설정된 한계 값의 진리결과를 플롯한다.
- ROC : Receiver Operating Characteristic : FPR(False Positive Rate), TPR(True Positive Rate)의 결과값을 플롯한다. roc_curve()
- AUC : Area Under the Curve(PR곡선). 세로측에서 적합률 / 가로측 재현률(TPR)
    -> 편차가 큰데이터를 분류했을경우 플롯 ex) 공장에서 기계고장 탐지. 일반기계(negative)
- ROC_curve
- ROC-AUC : roc_auc_score()
- ROC 곡선에서 모델을 평가할 때 일반적인 것이 AUC이다.
- AUC는 ROC 곡선의 아래 면적을 말하고 AUC가 높을수록 모델의 분류 성능이 높다(평가율이 좋다)
- 예측점수의 순서를 구현할 때 사용, 무작위 분류가 될 경우
- 예측점수란 예측된 확률을 말한다. 1.0 -> 100%의 확률 0.8 -> 80% 전체 기준을 1로 잡았을 경우
- 임계 값을 예측점수의 최소값이면 모두 1로 판정된다

```python
# 1. roc_curve() 함수를 사용해보자
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt

y_true = [0,0,0,0,1,1,1,1]
y_score = [0.2,0.3,0.6,0.8,0.4,0.5,0.7,0.9]

roc = roc_curve(y_true, y_score)
print(roc, type(roc), len(roc))
```

```python
fpr, tpr, thresholds = roc_curve(y_true, y_score)
print(fpr)
print(tpr)
print(thresholds)
# FPR = (TN + FP) / FP  # 오답 비율
# TPR = (TP + FN) / TP  # 정답 비율

plt.plot(fpr, tpr, marker='o')
plt.xlabel('FPR : False Positive Rate')  
plt.ylabel('TPR : True Positive Rate')   
plt.grid()
```

```python
fpr, tpr, thresholds = roc_curve(y_true, y_score, drop_intermediate = False)
print(fpr)
print(tpr)
print(thresholds)
# FPR = FP / (TN + FP)  # 오답 비율
# TPR = TP / (TP + FN)  # 정답 비율

plt.plot(fpr, tpr, marker='o')
plt.xlabel('FPR : False Positive Rate')  
plt.ylabel('TPR : True Positive Rate')   
plt.grid()
```

```python
from sklearn.metrics import roc_curve, recall_score, confusion_matrix
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

y_true = np.array([0,0,0,0,1,1,1,1])
y_score = np.array([0.2,0.3,0.6,0.8,0.4,0.5,0.7,0.9])

print(y_score >= 0.5, '\n')  # 예측률의 한계값을 0.5로 비교확인

print((y_score >= 0.5).astype(int))  # 정수로 0, 1로 바꾸자
```

```python
# 임의 함수
def fpr_score(y_true, y_pred):
    tn,fp,fn,tp = confusion_matrix(y_true, y_pred).flatten()
    return fp / (tn + fp)


print(fpr_score(y_true, y_score >= 0.5))  # 0.5

print(recall_score(y_true, y_score >= 0.5))  # 0.75
```

```python
th_min = min(y_score)
print(th_min)
print((y_score >= th_min).astype(int))
print(fpr_score(y_true, y_score >= th_min))
```

```python
th_max = max(y_score)
print(th_max)
print((y_score >= th_max).astype(int))
print(fpr_score(y_true, y_score >= th_max))
```

```python
df = pd.DataFrame({'True':y_true, 'score':y_score})
df['TPR'] = df.apply(lambda row: recall_score(y_true, y_score >= row['score']), axis=1)
df['FPR'] = df.apply(lambda row: fpr_score(y_true, y_score >= row['score']), axis=1)
df
```

```python
# 소수 판별

import math

def isPrime2(n):
    for i in range(2, math.floor(math.sqrt(n))+1):   # 루트n 까지만 확인
        if (n % i == 0):
            return False
    return True
```

```python
# 소수 판별
# 1보다 큰 자연수 n이 주어졌을 때, n보다 작은 소수의 개수를 구하는 함수를 만들어보자

def countPrimes(n):
    count = 0
    for i in range(2, n+1):
        if isPrime2(i):
            count += 1
    return count
```

```python
# 소수 판별
# 1보다 큰 자연수 n이 주어졌을 때, n보다 작은 소수의 개수를 구하는 함수를 만들어보자
import math

def isPrime3(n):
    primes = []
    
    for i in range(2, n):
        check = True
        for j in range(2, math.floor(math.sqrt(i))+1):
            if i % j == 0:
                check = False
                break
                
        if check:    
            primes.append(i)
        
    return len(primes)
```

```python
# 시간 측정

import time

n = 1000000
start = time.time()
print(countPrimes(n))
elapsed = time.time() - start   # 걸린 시간 = 끝 시간 - 시작 시간

print(elapsed, '초')
print(elapsed/60, '분')
print(elapsed/60/60, '시간')
```

```python
# 소수 판별 - 에라토스테네스의 체

def Eratosthenes(n):
    sieve = [False, False] + [True] * (n-1)  # 0과 1은 소수가 아니므로 기본적으로 False로 두고 나머지는 True로 초기값을 준다.
    primes = []   # 소수값이 들어갈 리스트
    
    for i in range(2, n+1):
        if sieve[i]:             # sieve 리스트의 i번째 값이 True. 즉 소수이면
            primes.append(i)     # primes 리스트에 추가
            
        for j in range(i*2, n+1, i):    # i값의 2배부터 n값까지 배수를 찾는다.
            sieve[j] = False            # i값의 배수들은 제거
            
    return len(primes)
```

```python
# 시간 측정 - 에라토스테네스의 체

import time

n = 1000000
start = time.time()
print(Eratosthenes(n))
elapsed = time.time() - start   # 걸린 시간 = 끝 시간 - 시작 시간

print(elapsed, '초')
print(elapsed/60, '분')
print(elapsed/60/60, '시간')
```

```python
# 소인수 분해

def factorize(n):
    factor = 2      # 시작 값
    factors = []    # 소인수를 담을 리스트
    
    while (factor**2 <= n):           # n까지 확인하지않고 n의 제곱근까지 확인
        while (n % factor == 0):      # factor의 값이 n의 소인수이면 루프
            n = n // factor            # n의 값을 변경해준다.
            factors.append(factor)     # 소인수이기 때문에 리스트에 담는다.
        factor += 1         # 소인수가 아니면 루프를 빠져나와 factor의 값을 1더한다.
    
    if n > 1:
        factors.append(n)   # 루프를 다 돌고 n의 값이 더이상 나눠지지 않는데 n이 1보다 크면 그 값도 소인수이기 때문에 추가
    
    return factors

factorize(45)
```

```python
# 소인수 찾기

def findFactors(n):
    factor_list = factorize(n)    # 사전 제작한 factorize 함수 사용
    factors = []                  # 소인수를 담을 리스트
    
    for i in factor_list:
        if (i not in factors):   # factors 리스트에 소인수 값이 없으면 추가
            factors.append(i)
            
    return factors

findFactors(3757208)
```

```python
# 소인수 찾기

def factorize(n):
    factor = 2      # 시작 값
    factors = []    # 소인수를 담을 리스트
    
    while (factor**2 <= n):           # n까지 확인하지않고 n의 제곱근까지 확인
        while (n % factor == 0):      # factor의 값이 n의 소인수이면 루프
            n = n // factor            # n의 값을 변경해준다.
            factors.append(factor)     # 소인수이기 때문에 리스트에 담는다.
        factor += 1         # 소인수가 아니면 루프를 빠져나와 factor의 값을 1더한다.
    
    if n > 1:
        factors.append(n)   # 루프를 다 돌고 n의 값이 더이상 나눠지지 않는데 n이 1보다 크면 그 값도 소인수이기 때문에 추가
    
    return list(set(factors))

factorize(45)
```

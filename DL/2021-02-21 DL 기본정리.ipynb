```python
* 편차 : 원점수에서 평균 뺀 값(원점수 174 - 평균 179 = 편차 -5). 편차를 다 더하면 0
* 분산: 자료와 평균과의 거리(⇒편차)제곱에 대한 중간값. 또는 거리의 중간값. 분포의 면적. 표준편차의 제곱 값. 편차제곱의 평균
* 표준편차 : 자료와 평균과의 거리. 평균으로 부터 값이 얼마만큼 떨어져있나. 분산의 제곱근

모집단 : 정보를 얻고자 하는 관심 대사의 전체 집합
모수 : 모집단의 분포가 가지는 특성값
표본 : 모집단의 부분집합. 전수 검사는 어렵기 때문에 일부 추출
  - 반드시 랜덤으로 뽑아야한다
  - 추출 할 때 마다 다른값으로 구성
  - 표본 통계량은 항상 오차가 수반 -> 표준 오차(표본 통계량의 표준편차)

* 표준정규분포 : 평균 0이고 분산 1인 분포
표준정규분포로 만드는것이 표준정규화 -> 평균을 빼주고 표준편차로 나눠준다
x1 - 평균 / 표준편차

* 프로그램 : 기계가 해야 할 일을 기계가 알아들을 수 있는 방식으로 순서대로 적은 것

* 데이터 과학 : 데이터를 만들고, 만들어진 데이터를 이용하는 일을 하는 것
* 데이터 공학 : 데이터를 다루는 도구를 만들고, 도구를 관리하는 일을 하는 것

* 행(가로줄)(row) : 개체(instance), 관측치(observed value), 기록(record), 사례(example), 경우(case)
* 열(세로줄)(column) : 특성(feature), 속성(attribute), 변수(variable), field

* 독립변수 : 원인이 되는 열. 결과에 영향을 받지 않는 독립적인 사건

* 종속변수 : 결과가 되는 열. 원인에 종속되어서 발생한 사건

* 상관관계 : 두 변수 들이 서로 관련, 상관이 있는 관계

* 인과관계 : 두 변수 들이 원인과 결과의 관계

상관관계는 인과관계를 포함. 상관관계>인과관계
 -> 두 변수가 관련, 상관이 있을 수 있지만 꼭 원인과 결과를 가진 관계가 아닐 수 있다.
```

```python
* 딥러닝
Input Layer(입력신호) -> Hidden Layer(은닉층) -> Output Layer(출력층)
Input Layer(입력신호) -> 가중치 -> 입력신호의 총합 + 바이어스 -> 활성화 함수 -> Output Layer(출력층)

최적의 가중치 값들을 찾는 과정

가중치(weight) : 입력신호의 강도를 표현. 앞에서 전달받은 정보를 얼마나 중점적으로 고려할 것인지
    
입력신호의 총합(summation) : 각 입력신호에 가중치를 곱하여 합한 값(w1*x1 + w2*x2 + ... + wn*xn)
    
바이어스(bias) : 사용자가 부여하거나 랜덤으로 초기화 되는 수. 입력된 신경망과 가중치 곱의 합이 가져야 할 최소 기준을 정의
                 최소 기준을 정의하고 그 기준을 넘어야 신호가 활성화. 
                 기준이 높으면 연결강도가 강한 신호만 살아남고, 낮으면 모든 신호들이 다 살아남는다. 적정한 기준을 잡는 것이 중요
        
활성화 함수(activation function) : 신호의 총합을 출력신호로 변환.
                                   입력신호의 총합과 바이어스 값을 받고 활성화 함수의 기준에 따라 활성화 여부를 결정
                                   단일 노드에 여러 신호가 들어오면, 다음 레이어(layer)에 보낼 신호의 강도를 결정.
                                   NN(Neural Network)에 비선형성을 추가하는 역할
                                   입력신호는 선형인데(선은 데이터를 정확히 분류하지 못함) 활성화함수를 사용해 비선형을 만든다. 
                                    -> 비선형적으로 정확히 분류한다.
```


입력된 데이터(X)를 통해 정답 후보 값인 Y’을 예측하여 실제 정답인 Y와 비교하여 차이 계산<br>
계산된 차이 만큼의 손실 값(loss score)을 활용하여 최적화 기법(Optimizer)으로 새롭게 업데이트할 가중치(weights) 계산<br>
가중치를 업데이트하여 입력 데이터 X에 대한 새로운 정답 후보 값 Y’ 예측

* 최적화기법(Optimizer) : 손실값을 가지고 가중치를 어떻게, 어떤값으로 업데이트 시킬지 적용하는 기법<br>
최적의 weight를 찾아 업데이트하기 위한 알고리즘

```python
epoch : train data set으로 한 번 학습을 완료한 상태
    ex) 10 epoch는 동일한 train data set으로 10번 학습한다.
    
batch size : train data set을 여러개로 나누어 데이터 셋을 만든다. 나눠진 데이터 하나를 batch(mini-batch)라고 하고
             batch size는 batch에 속하는 데이터 수를 의미
             전체 트레이닝 셋을 작게 나누는 이유는 트레이닝 데이터를 통째로 신경망에 넣으면 비효율적이고 리소스 사용으로 학습 시간이 오래 걸리기 때문
```

DNN(Deep Neural Network) : 딥러닝
학습(learning)은 최적의 가중치(weight)들의 조합을 찾는 과정

인공신경망의 훈련과정
1. 전방전파(Feed Forward) : input 값과 weight / bias를 이용한 예측 값 계산
<br>   

2. 오차계산(Calculate Loss) : 입력된 데이터(X)를 통해 정답 후보 값을 예측하여 실제 정답 값 비교하여 차이 계산
<br> 

3. 오차 역전파(Back Propagation) : Gradient(경사로) Descent(따라내려가다) 알고리즘을 이용한 weigt 계산. <br>발생한 오차만큼의 정보량을 뒤로 보내, 신경망이 가지고 있는 weights를 업데이트 하는 과정
<br>

4. 가중치 업데이트(Weight update) : 가중치 업데이트


* Gradient(기울기)
 - 손실 함수의 weight들에 대한 모든 기울기로 미분(편미분)한 값의 집합(벡터)
 - 가중치 매개변수의 값을 변화시켰을때 손실 함수의 변화량
    
    
* Gradient Descent

  - 가중치와 손실함수(loss function)의 관계를 이용하여 최적의 가중치를 찾는 과정<br>
  - 특정 지점 w에서 기울기가 가장 가파르게 하강하는 곳을 따라 learning rate(학습률)만큼 이동하면서 손실 값 최소 지점으로 이동해 가며 최적의 가중치를 찾는 알고리즘<br>
  - 기울기가 작아지는 지점으로 찾아가면서 0인 지점으로 찾아가는 과정
  
<br>

* learning rate(학습률)
  - Gradient(기울기)에 의해 결정된 이동방향에 따라 얼마만큼 이동 크기를 결정하기 위한 Hyper Parameter
  - 일반적으로 0.01 ~ 0.001 사이의 값을 가장 많이 사용
  
* loss function(손실함수)
  - 실제 값과 예측 값을 비교하여 차이를 계산하기 위한 함수

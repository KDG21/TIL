```python
* 텍스트 전처리 :  정제, 불용어 제거, 어간 추출, 토큰화 및 문서 표현 등을 진행하는 과정(pytorch 사용)

정제 : 특수문자 등과 같은 불필요한 노이즈 텍스트 제거 및 대소문자 통일
불용어 제거 : 전치사, 관사 등 문장이나 문서의 특징을 표현하는데 불필요한 단어를 제거하는 단계
어간추출 : 단어의 기본 형태를 추출하는 단계
토큰화(Tokenization) : 코퍼스(corpus)에서 분리자(separator)를 포함하지 않는 연속적인 문자열 단위로 분리(음소, 음절, 단어, 형태소 단위로 설정할 수 있음)
임베딩 과정에서 어느 단위로 토큰화할건지 필요
코퍼스 : 문서의 집합

* 음소 < 음절 < 형태소 < 어절 < 문장 -> 문서

* 형태소의 개념 : 언어에서 의미를 가지는 최소단위
* 형태소 분석의 개념 : 형태소 분석은 자연어 문장에서 의미를 가진 최소 단위인 형태소(명사, 동사, 형용사, 부사, 조사, 어미 등)를 분석
  자연어 처리를 위해 수행해야 하는 가장 첫 단계의 분석

* 임베딩 : 범주형 자료를 연속형 벡터 형태로 변환시키는것
워드 임베딩 : 인간의 언어를 컴퓨터가 이해할 수 있는 언어로 변환 -> 문자열을 숫자로 변환하여 벡터 공간에 표현

문장 단위 워드 임베딩(언어 모델)은 문맥을 이해할 수 있음
형태소 분석기는 문맥을 이해하지 못함

* 슬라이딩 윈도우
윈도우 사이즈가 2인 경우 2-gram

* 분산표현방법 : 비슷한 위치에서 등장하는 단어들은 비슷한 의미를 가질 확률이 높다는 가정하에 적용한 것
* Word2Vec(분산표현방법의 일종) -> Doc2Vec

* gensim 패키지
Word2Vec 단어와 유사한 단어를 찾을 수 있음
Doc2Vec 이용하면 문장의 유사도, 문맥을 이해할 수 있음


* 유사도 : 코사인 유사도, 자카드 유사도, 레벤슈타인(편집 거리)

* 거리 : 유클리디안 거리, 가우시안 거리, 맨하탄 거리

* dtm, tdm으로 특정 단어가 얼마나 나오는지 보고 문서의 유사성을 확인할 수 있음
```

```python
# exam)
import numpy as np
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('C:\\Users\\user\\MyPython\\img\\glasses02.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

glasses = cv2.CascadeClassifier('C:\\Users\\user\\Downloads\\opencv-master\\data\\haarcascades\\haarcascade_eye_tree_eyeglasses.xml')
body = glasses.detectMultiScale(gray, 1.01, 10)

for (x,y,w,h) in body:
    cv2.rectangle(gray, (x,y), (x+w,y+h), (255,0,0),3)

plt.figure(figsize=(12,12))
plt.imshow(gray, cmap='gray')
plt.xticks([]), plt.yticks([])
plt.show()
```
